```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('Universities.csv')
```

```python
df1
```

```python
plt.hist(df1['SAT'])
plt.xlabel('SATSCORES')
plt.ylabel('Frequency')
plt.title('histogram of SAT scores')
plt.show()
```

```python
sns.histplot(df1['SAT'])
plt.show()
```

```python
sns.histplot(df1['Accept'], color='brown')
plt.xlabel('Acceptance Rate')
plt.ylabel('Frequency')
plt.title('Histogram of Acceptance Rates')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Accept'],'bo')
plt.xlabel('SATSscore')
plt.ylabel('Acceptance Rate')
plt.title('Scatter Plot of SAT Scores vs. Acceptance Rates')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Top10'],'rs')
plt.xlabel('SATSscore')
plt.ylabel('Top 10 Universities')
plt.title('Scatter Plot of SAT Scores vs. Top 10 Universities')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Top10'],'o')
plt.xlabel('SATSscore')
plt.ylabel('Top 10 ')
plt.title('Scatter Plot of SAT Scores vs. Top 10 ')
plt.show()
```

```python
sns.pairplot(df1)
```

```python
sns.pairplot(df1.iloc[:,1:3])
plt.show()
```
